**Smart Student Performance Analyzer (NumPy Only)**

```python
#Each row = one student
#Each column = one subject
import numpy as np

marks = [
    [85, 70, 90], # student 1 
    [60, 75, 65], #student 2 
    [95, 88, 92],
    [40, 55, 60]
]
```

```python
# convert python list into array:
marks = np.array(marks)
marks
```

```python
#Find average of each student
count = 1
for i in range (len(marks)):
    print(f" for student {count} marks are {marks[i].mean()}")
    count+=1
```

```python
#Subject-wise average
count =1
for i in range(len(marks)-1):
    print(f"student {count} average marks {marks[:,i].mean()}")
    count+=1
```

```python
# Find who got the highest total marks
total = marks.sum(axis=1)
higest_marks = total.argmax()

print(f" Higest marks gain by student {higest_marks +1}")

```

```python
#Failed students
#If average < 60 → fail
avg = marks.mean(axis=1)

b = avg <60

print(f"failed student {marks[b]}")
```

```python
#Rank students
total_mark = np.sum(marks,axis =1)

rank = np.argsort(total_mark)
rank_2 =rank+1
rank_2
```
