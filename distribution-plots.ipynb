```python

import seaborn as sns
import matplotlib.pyplot as plt
```


**We use distribution plots to visualize the distribution of quantitative data.Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:
* distplot/histplot
* jointplot
* pairplot
* rugplot

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df["size"].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins =20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins =20,kde = True)
```

**Joinplot**

```python
sns.jointplot(x="total_bill",y="tip",data= df,kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data= df,kind = "kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data= df,kind = "hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data= df,kind = "reg")
```

**pairplot**

```python
sns.pairplot(df,hue="size",palette="rainbow")
```

**rugplot**

```python
sns.rugplot(df["total_bill"])
```

```python
sns.rugplot(df)
```
