```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basic

```python
x = np.linspace(0,5,10)
y = x**2
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Subplot in matplot lib 

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
```

**Objext oriented Way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig=plt.figure(figsize = (12,8),dpi= 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 =fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```
